```python
inputs = tf.placeholder(tf.float32, shape=[2, 299, 299, 3])
```

```python
with slim.arg_scope(resnet_utils.resnet_arg_scope()):
# with slim.arg_scope(resnet_v1_beta.resnet_arg_scope()):
        _, end_points = resnet_v1_beta.resnet_v1_101_beta(inputs, num_classes=1001, is_training=False, reuse=tf.AUTO_REUSE)
```

```python
saver = tf.train.Saver()
sess = tf.Session()
saver.restore(sess, './resnet_v1_101/model.ckpt')
```

```python
resnet_pt = []
with open('./pytorch_res.txt', 'r') as f:
    for line in f:
        print(line.strip())
        resnet_pt.append(line.strip())
```

```python
tvs = [v for v in tf.global_variables()]
for v, k in zip(tvs, resnet_pt):
    print(v.name)
    print(k)
    #print(v.read_value())
    output = sess.run(v)
    print(output.shape)
    print('\n')
```

```python
rst = {}
for pt_k, tf_v in zip(resnet_pt, tvs):
    print(tf_v.name)
    print(pt_k)
    output = sess.run(tf_v)
    rst[pt_k] = output
    print(output.shape)
    print('\n')
    
```

```python
with open('res101.npy', 'wb') as f:
    np.save(f, rst)
```
