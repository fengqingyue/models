```python
import numpy as np
import tensorflow as tf

from core import xception
from tensorflow.contrib.slim.nets import resnet_utils
```

```python
slim = tf.contrib.slim
```

```python
inputs = tf.placeholder(tf.float32, shape=[2, 299, 299, 3])
```

```python
with slim.arg_scope(xception.xception_arg_scope()):
# with slim.arg_scope(resnet_v1_beta.resnet_arg_scope()):
        _, end_points = xception.xception_65(inputs, num_classes=1001, is_training=False, reuse=tf.AUTO_REUSE)
```

```python
saver = tf.train.Saver()
sess = tf.Session()
saver.restore(sess, './xception/model.ckpt')
```

```python
tvs = [v for v in tf.global_variables()]
for v in tvs:
    print(v.name)
```

```python
xception_name = []
with open('./pytorch_xception.txt', 'r') as f:
    for line in f:
        print(line.strip())
        xception_name.append(line.strip())
```

```python
rst = {}
for name, val in zip(xception_name, tvs):
    print(name)
    print(val.name)
    output = sess.run(val)
    print(output.shape)
    rst[name] = output
    print('\n')
```
